```python
# Data Structures

# List -> store continous or discrete values! of same or different types
# representation ->    [   ]
# ORDERED, INDEXED -> a[5], a[0], mylist[-3] -> 5 or 0 or -3 were indexes

# Tuple -> READ ONLY
# representation ->     (     )
# ORDERED, INDEXED

# Set -> { apple, banana } or { banana, apple} or {apple, banana, apple, apple} = SAME
# No duplication, ORDER does not matter
# representation ->     {      }
# UNORDERED, UNINDEXED -> you cannot query a set like myset[3], yourset[-1]

# Dictionary -> key-value pairs, key and value can be anything
# representation ->    {  keys : values        }
# UNORDERED, INDEX with keys!
```

```python
mylist = [ 'Bengaluru', 'Chennai', 'Hyd', 'Noida', 'Delhi']
myHousePriceList = [1000, 2000, 1500, 1400, 2500]

```

```python
#dictionary of city-price
cityprice = {
    'City Name': mylist,
    'price': myHousePriceList
}
print(cityprice)
```

```python
# only possible if all columns have same lenght-> then you can convert dictionary into a DATASET
import pandas as pd
cityprice_dataframe = pd.DataFrame(cityprice)
print(cityprice_dataframe)
```

```python
# lambda functions are 1-line execution!
# lambda functions unlike traditional function calls DO NOT MAKE A MEMORY JUMP and are executed
# in-place in-memory!

# syntax -> lambda input: what to do with input

# syntax-> square = lambda x : x*x

price_logic_for_usd = lambda price: price / 80.0

cityprice_dataframe['price_in_usd'] = cityprice_dataframe['price'].apply(price_logic_for_usd)
cityprice_dataframe
```

```python
# ** -> to the power

def mypower(yourpower):
  return lambda base: base ** yourpower

square = mypower(2) # lambda base: base ** 2
cube = mypower(3) # lambda base: base ** 3

print(square(3)) # lambda 3: 3 ** 2
print(cube(9)) # lambda 9: 9 ** 3
```
